```
import pandas as pd
import numpy as np
from numpy.random import randn
np.random.seed(101)
df = pd.DataFrame(randn(4,3),index='A B C D'.split(),columns='W X Y'.split()) # Creating column
df
```

```
df['W'] # Showing column W
df[['W','Y']] # Pass a list of column names only
```

```
df['new'] = df['W'] + df['Y'] # addition of two arrays
df
```

```
df['avg'] = (df['W'] + df['X'] +df['Y'])/3 #Finding average
df
```

```
df.drop('new',axis=1) #Removing Column
```

```
df.drop('avg',axis=1,inplace=True) #dropping avg column from the table
df
```

```
df.drop('A',axis=0) # displaying result without row A
```

```
df.loc['B'] # location of column B
```

```
df.iloc[2] #index location
```

```
df.loc[['B','C'],['W','Y']] #location of [B C] , [W Y]
```

```
df>0 #Result in true or false of df greater than 0
```

```
df[df>0] # Values greater than 0
```

```
df[df>1] # values greater than 1
```

```
df[df['W']>-1] #column value greater than -1
```

```
df[df['W']>0]['Y'] #Finding vlaue W > 0 * Y
```
