TOP 20 COUNTRIES ACCORDING TO THE NO OF CONFIRMED CASES

```
df1=pd.read_csv("/content/drive/My Drive/data analytics dataset/time-series-19-covid-combined.csv")
```

```
print(df1.columns)
print(df1.dtypes)
print(df1)
```

```
df2=df1.groupby("Country/Region").last()
```

```
df3=df2.sort_values("Confirmed",ascending=False).head(20)
print(df3["Confirmed"].astype(int))
```

TIME SERIES OF THE COUNTRIES WITH THE MOST CONFIRMED CASES

```
df3["Countries"]=df3.index
df3.plot(kind="bar",x="Countries",y="Confirmed")
```

SHOW TOP 20 COUNTRIES ACCORDING TO THE NO OF DEATHS

```
df4=df2.sort_values("Deaths",ascending=False).head(20)
print(df4["Deaths"].astype(int))
```

```
df4["Countries"]=df4.index
df4.plot(kind="bar",x="Countries",y="Deaths")
```
